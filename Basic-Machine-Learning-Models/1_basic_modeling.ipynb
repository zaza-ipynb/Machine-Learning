# Create a Basic Machine Learning Model to predict a value

#1. Import Library and Data

```
# import libraries 
import tensorflow as tf 
import tensorflow.keras as keras
import numpy as np
```

Take a look at this pair of data and labels(result). We are going to make the machine learns to map `x` to `y`. 

Because `y` is a continous number, this task is considered as "regression" task

```
# create an example of train data
x = np.array([1,2,3,4,5]) # data 
y = np.array([3,5,7,9,11]) # target label 
```

If you pay attention to the pair of x and y, you will find that $y = 2x + 1$. 

We will see if the machine are capable in estimating the y value out of x. 

___

#2. Create the Model

```
# create a simple model
model = keras.Sequential()
model.add(keras.layers.Dense(units=1, input_shape=[1]))
model.compile(optimizer='SGD', loss='mean_squared_error')
```

You don't have to worry about the codes, but here's what your model will look like. 
Mathematically, you will have a model of $y = wx+b$

image.png

#3. Train the Model

```
# train the model
model.fit(x,y, epochs=100)
```

```
# print the value of w and b of your model 
print('the w value of current model is: ', model.get_weights()[0][0])
print('the b value of current model is: ', model.get_weights()[1])
```

#4. Test the Model

```
# use the model to predict some values
model.predict([1])
```

There you have it, your machine learning model that are able to predict $y = 2x+1$
